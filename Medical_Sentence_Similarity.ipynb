##BERT Similarity

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer('bert-base-nli-mean-tokens')
```

```python
sentence_embeddings = model.encode(sentences)
```

```python
sentence_embeddings.shape
```

```python
similarity=[]
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
for i in range(0,len(xx)):
  new=[xx[i]]
  new_sec = model.encode(new)
  tt=cosine_similarity([new_sec[0]],sentence_embeddings[0:])
  similarity.append(tt)
```

```python
len(similarity[0][0])
```

```python
import numpy as np
```

```python
aa=pd.DataFrame(index=np.arange(len(similarity[0][0])), columns=np.arange(0))
```

```python
aa.insert(loc=0, column='Title', value="NaN")
```

```python
aa["Title"]=sentences
```

```python
xx[1]
```

```python
for i in range (0,len(aa)):
  try:
    aa.insert(i+1,xx[i],similarity[i][0])
  except IndexError:
    pass
  continue
```

```python
aa
```

```python
for i in range(1,len(xx)+1):
  aa.iloc[:,i] = pd.Series(["{0:.2f}% string match".format(val * 100) for val in aa.iloc[:,i]], index = aa.index)
```

```python
aa
```

```python
cause_per=[]
```

```python
for i in range(0,255):
  dd=i+1
  rr="Cause"
  rrr="{}{} % Match".format(rr,dd)
  cause_per.append(rrr)
cause_per.insert(0,"Public Title of Study")
```

```python
aa.columns=cause_per
```

```python
cause=[]
```

```python
for i in range(0,255):
  dd=i+1
  rr="Cause"
  rrr="{}{}".format(rr,dd)
  cause.append(rrr)
```

```python
aa
```

```python
len(xx)
```

```python
lst = list(range(1,510,2))
```

```python
len(lst)
```

```python
for i in range (0,len(xx)):
  aa.insert(lst[i],cause[i],xx[i])
```

```python
aa
```

```python
aa.to_excel("final_phara_2_bert.xlsx",index=False,
             sheet_name="Match string")
```

##TF-IDF Similarity

```python

import nltk
nltk.download('punkt')



from nltk.tokenize import word_tokenize
gen_docs = [[w.lower() for w in word_tokenize(text)] 
            for text in raw_documents]
print(gen_docs)


dictionary = gensim.corpora.Dictionary(gen_docs)

corpus = [dictionary.doc2bow(gen_doc) for gen_doc in gen_docs]
print(corpus)




tf_idf = gensim.models.TfidfModel(corpus)
print(tf_idf)
s = 0
for i in corpus:
    s += len(i)
print(s)

sims = gensim.similarities.Similarity('/content/sample_data',tf_idf[corpus],
                                      num_features=len(dictionary))
print(sims)
print(type(sims))
```

```python
dictionary.token2id
```

```python
similarity=[]
```

```python
for i in range(0,len(xx)):
  query_doc = [w.lower() for w in word_tokenize(xx[i])]
  query_doc_bow = dictionary.doc2bow(query_doc)
  query_doc_tf_idf = tf_idf[query_doc_bow]
  sim1=sims[query_doc_tf_idf]
  new_list_1 = [ str(round(x*100))+"% string match" for x in sim1]
  similarity.append(new_list_1)
```
